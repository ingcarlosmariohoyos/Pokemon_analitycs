**Este notebook tiene como objetivo realizar la limpieza general del dataset**

```python
import pandas as pd

df = pd.read_csv('../data/raw/pokemon.csv')

print(df)
```

```python
#En el notebook anterior se evidencio ue el campo tipo 2 presentaba nulos por lo cual se procede a dar el manejo de estos campos
#Generamos un diccionario que toma como clave el campo Type 2 y como valor la palabra no aplica para sustituir los campos vacios
valor_tipo = {"Type 2":"No aplica"}
#Generamos en dataframe nuevo el cual va aplicar la funcion fillna para reemplazar los campos vacios
df_sin_na = df.fillna(valor_tipo)
df_sin_na
```

```python
#comprobamos que no queden campos vacios aplicando la funcion isnull y sumando en cada columna cuantos nulos exites en este caso 0
df_sin_na.isnull().sum()
```

```python
#validamos los campos existentes nuevamente en nuestro dataset
df.info()
```

```python
# En este caso se genero otro diccionario llamado columnas donde la clave sera el nombre de la columna actual 
# y el valor sera el nuevo nombre que le daremos a la columna 
Columnas = { 
   '#': 'Numero Dex',
   'Name': 'Nombre',
   'Type 1': 'Tipo Primario',
   'Type 2': 'Tipo Secundario',
   'Total': 'Estadisticas Totales',
   'HP': 'Puntos Salud',
   'Attack': 'Ataque',
   'Defense': 'Defensa',
   'Sp. Atk': 'Ataque Especial',
   'Sp. Def': 'Defensa Especial',
   'Speed': 'Velocidad',
   'Generation': 'Generacion',
   'Legendary': 'Legendario'
}
# creamos un nuevo dataframe que contendra los datos del dataset pero con el nuevo nombre de las columnas mas claros y para mayor entendimiento
pokedex = df_sin_na.rename(columns = Columnas)
pokedex

```

```python
# Ahora se van a categorizar el tipo de fase en la que se encuentra el pokemon primero se revisaran la cantida de megaevoluciones
es_mega = pokedex['Nombre'].str.contains("mega", case = False)
# generamos un dataset temporal que nos mostrara solo los pokemon que son megaevoluciones
megas = pokedex[es_mega]
print(f"Total de Pokémon Mega encontrados: {len(megas)}")
```

```python
# Ahora vamos a sacar los que no son megaevoluciones y dejaremos los que estan en su estado base realizando una negacion en base a la variable
# anterior que nos indicaba cual era megaevolucion
no_es_mega = ~es_mega
# En base a esta variable que nos indica cuales no son  megas sacamos la cantidad de pokemon en estado base
pokedex_base = pokedex[no_es_mega]
print(f"Total de Pokémon Base (no Mega) encontrados: {len(pokedex_base)}")
```

```python
# El resultado de el dataframe base deja como evidencia solo los pokemon en estado base
pokedex_base
```

```python
# El resultado de el dataframe mega da como resultado los pokemon con megaevoluciones pero con un ligero error
# Se evidencia que el pokemon Meganium quedo agregado a el dataframe mega cuando deberia estar en el base se corregira en lineas posteriores
megas.head(10)
```

```python
# Para corregir el error anterior se crea una variable Meganium en donde el datframe buscara por la columna nombre una coincidencia con el nombre del 
# Pokemon
Meganium = megas['Nombre'] == 'Meganium'
# creamos una nueva variable que tomara el resultado de la validacion anterior y la tendra a parte para insertarlo en el dataframe correcto
mover_pokemon = megas[Meganium]
# Ahora generaremos nuevamente el dataset con todas las filas que no contenan el nombre Meganium
Megas = megas[~Meganium]
```

```python
# Ahora podemos ver que ya no esta el pokemon Meganium en el dataframe megas
Megas
```

```python
# por medio de la funcion concat anexamos la fila que tiene los datos del pokemon Meganium a el datframe pokedex_base
Pokedex_Base = pd.concat([pokedex_base, mover_pokemon])
# reseteamos los indices de las filas para ubicar el registro en el nivel que pertenece
Pokedex_Base = Pokedex_Base.reset_index(drop=True)
#hacemos la validacion rapida para ver si quedo agregado al dataframe
Filtro_Pokemon = Pokedex_Base["Nombre"] == "Meganium"
Mostrar_Pokemon = Pokedex_Base[Filtro_Pokemon]
print(Mostrar_Pokemon)
```

```python
# procedemos a organizar de nuevo el dataframe por la columna Numero Dex que tiene el numero en el que deben ser ubicados
# Generamos un dataframe nuevo llamado Pokedex_Base_Ordenado
Pokedex_Base_Ordenado = Pokedex_Base.sort_values(by='Numero Dex', ascending=True)
Pokedex_Base_Ordenado = Pokedex_Base.reset_index(drop=True)
Pokedex_Base_Ordenado
```

```python
#Ordeno el df por el numero dex
Pokedex_Base_Ordenado = Pokedex_Base.sort_values(by='Numero Dex', ascending=True)

#reinicio el index del data frame
Pokedex_Base_Ordenado = Pokedex_Base_Ordenado.reset_index(drop=True)

#reviso si el pokemon quedo en su posicion correcta
print("Verificando la posición de Meganium (Número 154):")
# Mostramos las filas alrededor del 154
# Evidenciamos que quedo en su numero correcto
print(Pokedex_Base_Ordenado.loc[150:160, ['Nombre', 'Numero Dex']])
```
