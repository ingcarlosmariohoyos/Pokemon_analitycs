```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Pokedex_Base = pd.read_parquet('../data/processed/Pokedex_Base.parquet')
Megaevoluciones = pd.read_parquet('../data/processed/Megaevoluciones.parquet')
Top_Pokemon = pd.read_parquet('../data/processed/Top_Pokemon.parquet')
Top_Megaevoluciones = pd.read_parquet('../data/processed/Top_Pokemon_Megas.parquet')
```

**Este grafico nos permite mirar los 10 pokemon mas fuertes y este grafico de barras nos ayuda a verlos de manera ordena
En este caso podemos ver que el poder esta muy parejo entre la mayoria  del top 10 donde el unico relevante es slaking
podemos inferir que la cantidad de poder entre uno y otro no es tan importante por lo cual lo que define en este caso es el tipo de cada
pokemon**

```python
# Configuración básica para mejor la estética del grafico
sns.set_theme(style="whitegrid")

# 1. Definimos el tamaño del gráfico
plt.figure(figsize=(10, 6)) 

# 2. Crear el gráfico de barras horizontal (sns.barplot)
#    - Eje X: El valor numérico a comparar (Estadísticas Totales).
#    - Eje Y: La categoría (Nombre del Pokémon).
ax = sns.barplot(
    x='Estadisticas Totales', 
    y='Nombre', 
    data=Top_Pokemon, 
    hue='Nombre',        
    palette="viridis",
    legend=False         
)

# 3. Añadimos título y etiquetas
plt.title('Top 10 Pokémon por Estadísticas Totales', fontsize=16)
plt.xlabel('Estadísticas Totales', fontsize=12)
plt.ylabel('Nombre del Pokémon', fontsize=12)

#  Mostramos el valor exacto al final de cada barra (muy útil en Top N)
for i in ax.containers:
    # Mostramos el valor entero (sin decimales)
    ax.bar_label(i, fmt='%.0f', padding=5)

plt.tight_layout()
plt.show()
```

**Este grafico de cajas nos permite comparar el poder general entre regiones en donde evidenciamos que nos genero un outliner
en la region de Hoen ya que esta region posee criaturas con un poder superior a los 700 lo cual nos genera registros atipicos
por otro lado podemos ver que la region Shinoh presenta el mayor indice de poder general entre regiones**

```python
# Configuración básica para mejor la estética del grafico
sns.set_theme(style="whitegrid")

# 1. Definir el tamaño del gráfico
plt.figure(figsize=(12, 6))

# 2. Creamos el Box Plot (sns.boxplot)
#    - Eje X: La variable categórica (Generación).
#    - Eje Y: La variable continua (Estadisticas Totales).
ax = sns.boxplot(
    x='Region', 
    y='Estadisticas Totales', 
    data=Pokedex_Base,
    hue='Region',          
    palette="pastel",
    legend=False
)

# 3. Añadimos título y etiquetas
plt.title('Distribución de Estadísticas Totales por Generación', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Estadísticas Totales (Total Stats)', fontsize=12)

plt.tight_layout()
plt.show()
```

**Este grafico de histograma superpuesto nos muestra que hay una mayor variablilidad de estadisticas en los pokemon que no son legendarios
mientras que los pokemon legendarios presentan menos variabilidad y su indice de poder es mar cercano entre pokemones legendarios.**

```python
# Configuración básica para mejor la estética del grafico
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 7))

# Creamos el histograma con 'hue' para diferenciar Legendario y No Legendario
sns.histplot(
    data=Pokedex_Base, 
    x='Estadisticas Totales', 
    hue='Legendario',    # Usamos la columna 'Legendario' para diferenciar los grupos
    bins=30,             # Número de barras del histograma
    kde=True,            # Añadimos una estimación de la densidad del kernel (línea suavizada)
    alpha=0.5,           # Transparencia para ver ambas distribuciones
    palette="coolwarm"   # Paleta de colores para Legendario/No Legendario
)

plt.title('Distribución de Estadísticas Totales: Legendarios vs. No Legendarios', fontsize=16)
plt.xlabel('Estadísticas Totales', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend(title='Es Legendario') # Añadir una leyenda clara
plt.tight_layout()
plt.show()
```

```python
# Para poder crear la matriz de correlacion debemos realizar el siguiente paso
# Definimos las columnas de interés en este caso la de cada estadistica
columnas_stats = [
    'Puntos Salud', 'Ataque', 'Defensa', 
    'Ataque Especial', 'Defensa Especial', 'Velocidad'
]

# Creamos un sub-DataFrame solo con estas columnas
Estadisticas = Pokedex_Base[columnas_stats]
```

```python
matriz_correlacion = Estadisticas.corr()
print("Matriz de Correlación de Pearson:")
print(matriz_correlacion.round(2)) 
```

**El Mapa de Calor de Correlación reveló que, en general, las estadísticas presentan una correlación positiva, lo cual es esperable debido al factor de poder total del Pokémon (es decir, los Pokémon más fuertes tienen todas las estadísticas más altas). Sin embargo, identificamos una correlación negativa sutil pero significativa (cercana a $-0.3$) entre la Velocidad y las estadísticas defensivas (Defense / HP), lo que confirma la estrategia de balance de juego intencional que compensa la rapidez con la fragilidad, y viceversa.**

```python
# Configuración básica para mejor la estética del grafico
plt.figure(figsize=(9, 7))

# Crear el Heatmap
sns.heatmap(
    matriz_correlacion, 
    annot=True,     # Mostrar los valores de correlación en las celdas
    cmap='coolwarm',# Elegir una paleta de colores divergente (rojo/azul)
    fmt=".2f",      # Formato de dos decimales
    linewidths=.5,  # Líneas entre las celdas
    cbar=True       # Mostrar la barra de color
)

plt.title('Matriz de Correlación de Estadísticas de Pokémon', fontsize=14)
plt.show()
```

**"Este gráfico valida nuestra conclusión de correlación positiva, mostrando que el poder es interdependiente. Sin embargo, su mayor valor es la segmentación: la nube naranja de los Legendarios está desplazada hacia el cuadrante de alto poder, lo que verifica estadísticamente su estatus de élite. La variabilidad de los Legendarios en el cuadrante superior (ej. un Legendario con Ataque 150 y Defensa 50 vs. otro con Ataque 100 y Defensa 150) sugiere que, si bien son poderosos, conservan roles de combate bien definidos y compensatorios.**

```python
# Configuración básica para mejor la estética del grafico
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 8))

# Creamos el Gráfico de Dispersión (Scatter Plot)
ax = sns.scatterplot(
    x='Ataque',
    y='Defensa',
    data=Pokedex_Base,
    hue='Legendario',  # Usa el estado Legendario para colorear los puntos
    palette={False: '#6A9BC2', True: '#F08030'}, # Azul para No Legendario, Naranja para Legendario
    s=80,              # Tamaño de los puntos
    alpha=0.7          # Transparencia
)

# Añadimos títulos y etiquetas
plt.title('Relación entre Ataque y Defensa (Legendarios Destacados)', fontsize=16)
plt.xlabel('Puntos de Ataque', fontsize=12)
plt.ylabel('Puntos de Defensa', fontsize=12)

# Añadimos una línea de tendencia
sns.regplot(
    x='Ataque',
    y='Defensa',
    data=Pokedex_Base,
    scatter=False, # No queremos los puntos, solo la línea de regresión
    color='gray',
    ci=None        # No mostrar el intervalo de confianza
)

plt.legend(title='Es Legendario')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

**Podemos observar que los tipos primarios no estan bien equilibrados ya que no existen cantidades parecidas lo cual en la parte de tipos el 
juego esta desequilibrado para unos tipos mas que para otros**

```python
# Configuración básica para mejor la estética del grafico
sns.set_theme(style="whitegrid")

# Ajustar el tamaño de la figura para que quepan todos los Tipos
plt.figure(figsize=(14, 7))

# Creamos el Gráfico de Barras de Conteo
# 'Tipo Primario' en el eje X, Seaborn automáticamente calcula el conteo en el eje Y
ax = sns.countplot(
    y='Tipo Primario', # Usamos 'y' para barras horizontales, que son mejores para nombres largos
    data=Pokedex_Base,
    order=Pokedex_Base['Tipo Primario'].value_counts().index, # Ordenar las barras de mayor a menor
    palette='Spectral' # Puedes elegir la paleta de colores que prefieras
)

# Añadimos etiquetas y título
plt.title('Distribución de Pokémon por Tipo Primario', fontsize=16)
plt.xlabel('Cantidad de Pokémon', fontsize=12)
plt.ylabel('Tipo Primario', fontsize=12)

plt.tight_layout()
plt.show()
```
