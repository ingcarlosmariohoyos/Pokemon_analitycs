```python
import numpy as np
import pandas as pd
# Para esta ocasion decidi darle nombre mas especificos a los dataframe
Pokedex_Base = pd.read_parquet('../data/processed/Pokedex_Base_Ordenado.parquet')
Megaevoluciones = pd.read_parquet('../data/processed/Megas.parquet')
```

```python
# Validamos el correcto cargue de el dataset en el dataframe
Pokedex_Base
```

```python
# Validamos el correcto cargue de el dataset en el dataframe
Megaevoluciones
```

```python
# Dar nombre a las regiones con el fin de que podamos tener una nueva columna que indique el nombre de cada generacion
# Nos valemos nuevamente de los diccionarios para generar el nombre de las regiones
mapeo_regiones = {
    1:'Kanto',
    2:'Jhoto',
    3:'Hoen',
    4:'Shinoh',
    5:'Unova',
    6:'Kalos',
}
#Se crea una columna llamada region y se le asigna el valor de clave valor del diccinario para darle nombre a las regiones
Pokedex_Base['Region'] = Pokedex_Base['Generacion'].map(mapeo_regiones)
Pokedex_Base
```

```python
# para saber el numero de criaturas que hay por region utilizamos groupby por la nueva columna region y un count
# para saber la cantidad por region con el numero dex el cual representa un numero unico para cada pokemon 
conteo_region = Pokedex_Base.groupby('Region')['Numero Dex'].count()
print("\n Conteo de PokÃ©mon por RegiÃ³n:")
print(conteo_region)
```

```python
# para generar un dataframe de kanto primero de genera un filtro por region
Filtro_kanto = Pokedex_Base['Region'] == 'Kanto'
#luego se crea el dataframe
Pokedex_Kanto = Pokedex_Base[Filtro_kanto]
Pokedex_Kanto
```

```python
#para generar un dataframe  primero se genera un filtro por region
Filtro_Jhoto = Pokedex_Base['Region'] == 'Jhoto'
#luego se crea el dataframe
Pokedex_Jhoto = Pokedex_Base[Filtro_Jhoto]
Pokedex_Jhoto
```

```python
#para generar un dataframe  primero se genera un filtro por region
Filtro_Hoen = Pokedex_Base['Region'] == 'Hoen'
#luego se crea el dataframe
Pokedex_Hoen = Pokedex_Base[Filtro_Hoen]
Pokedex_Hoen
```

```python
#para generar un dataframe  primero se genera un filtro por region
Filtro_Shinoh = Pokedex_Base['Region'] == 'Shinoh'
#luego se crea el dataframe
Pokedex_Shinoh = Pokedex_Base[Filtro_Shinoh]
Pokedex_Shinoh
```

```python
#para generar un dataframe  primero se genera un filtro por region
Filtro_Unova = Pokedex_Base['Region'] == 'Unova'
#luego se crea el dataframe
Pokedex_Unova = Pokedex_Base[Filtro_Unova]
Pokedex_Unova
```

```python
#para generar un dataframe  primero se genera un filtro por region
Filtro_Kalos = Pokedex_Base['Region'] == 'Kalos'
#luego se crea el dataframe
Pokedex_Kalos = Pokedex_Base[Filtro_Kalos]
Pokedex_Kalos
```

```python
# Ahora vamos a realizar una categorizacion de el rol que ocupa cada pokemon 

# Atacante FÃ­sico
cond_atk = (Pokedex_Base['Ataque'] > Pokedex_Base['Ataque Especial']) & \
           (Pokedex_Base['Ataque'] > Pokedex_Base['Defensa']) & \
           (Pokedex_Base['Ataque'] > Pokedex_Base['Defensa Especial'])

# Atacante Especial
cond_sp_atk = (Pokedex_Base['Ataque Especial'] > Pokedex_Base['Ataque']) & \
              (Pokedex_Base['Ataque Especial'] > Pokedex_Base['Defensa']) & \
              (Pokedex_Base['Ataque Especial'] > Pokedex_Base['Defensa Especial'])

# Tanque (FÃ­sico o Especial)
cond_tank_fis = (Pokedex_Base['Defensa'] >= 100) & (Pokedex_Base['Defensa'] > Pokedex_Base['Defensa Especial'])
cond_tank_esp = (Pokedex_Base['Defensa Especial'] >= 100) & (Pokedex_Base['Defensa Especial'] > Pokedex_Base['Defensa'])

# Veloz (*Sweeper*)
# Calculamos el promedio de stats para comparar la velocidad
Pokedex_Base['Avg_Stats'] = Pokedex_Base[['Puntos Salud', 'Ataque', 'Defensa', 'Ataque Especial', 'Defensa Especial', 'Velocidad']].mean(axis=1)
cond_speed = (Pokedex_Base['Velocidad'] >= 100) & (Pokedex_Base['Velocidad'] > Pokedex_Base['Avg_Stats'])

# 5. Muralla (Alto HP y defensa)
cond_wall = (Pokedex_Base['Puntos Salud'] >= 100) & (Pokedex_Base['Defensa'] + Pokedex_Base['Defensa Especial'] >= 200)

# Definimos las opciones de rol en orden de prioridad
conditions = [cond_wall, cond_tank_fis, cond_tank_esp, cond_speed, cond_atk, cond_sp_atk]
choices = ['Muralla', 'Tanque FÃ­sico', 'Tanque Especial', 'Veloz', 'Atacante FÃ­sico', 'Atacante Especial']
```

```python
# Usamos np.select para aplicar la lÃ³gica. 'Balanceado' es el valor por defecto (default).
Pokedex_Base['Rol Pokemon'] = np.select(conditions, choices, default='Balanceado')
```

```python
Pokedex_Base
```

```python
# tambien generamos una nueva columna que sera la encargada de mostrarnos un promedio de las estadisticas de cada pokemon y lo redondeamos a un decimal
Pokedex_Base['Avg_Stats'] = Pokedex_Base['Avg_Stats'].round(1)
```

```python
# Renombramos la columna por un nombre mas entendible
Columna_promedio = {
    'Avg_Stats':'Promedio'
}
Pokedex_Base = Pokedex_Base.rename(columns = Columna_promedio)
```

```python
Pokedex_Base
```

```python
# Ahora se va a generar un dataframe con los 10 pokemon con mejores estadisticas
# primero vamos a excluir los legendarios de este dataframe de pokemon base por lo cual generamos un filtro

Filtro_legendarios = Pokedex_Base['Legendario'] == False
# Generamos un nuevo dataframe que contenga los pokemon sin legendarios
Pokedex_Sin_Legendarios = Pokedex_Base[Filtro_legendarios]

# Procedemos a realizar el ordenamiento por la columna estadististicas totales de mayor a menor e indicamos que solo nos muestre los primeros 10
Top_Pokemon = Pokedex_Sin_Legendarios.sort_values(by='Estadisticas Totales', ascending=False).iloc[0:10] 
Top_Pokemon
```
